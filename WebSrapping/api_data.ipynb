Extract movies from all the pages 

```python
df = pd.DataFrame()
```

```python
for i in range(1, 429):
    res = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=89abaea16d98800a6044f5fc706228d3&language=en-US&page={}".format(i))
    df_temp = pd.DataFrame(res.json()['results'])[['id', 'title', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    df = pd.concat([df, df_temp], axis = 0, ignore_index= True)
```
